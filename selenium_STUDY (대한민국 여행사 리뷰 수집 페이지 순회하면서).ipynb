```python
cnt = 300  # 예시로 100건의 리뷰를 수집하고자 할 때
page_cnt = math.ceil(cnt / 20)  # 리뷰 페이지당 20개의 리뷰가 있다고 가정

reviews_df = pd.DataFrame(columns=['리뷰 번호', '리뷰 내용'])  # 리뷰를 저장할 DataFrame

no = 0  # 수집된 리뷰의 개수를 카운트

for x in range(1, page_cnt + 1):
    print(f'========= {x} 페이지 작업 =========')
    # '_2L3vDiadT9' 클래스를 가진 모든 요소 찾기 (리뷰 내용을 포함)
    #review_elements = driver.find_elements(By.CSS_SELECTOR, 'div._3z6gI4oI6l > div > span')
    review_elements = driver.find_elements(By.CSS_SELECTOR, 'div._3z6gI4oI6l > div > span:last-child')


    for element in review_elements:
        no += 1
        if no > cnt:
            break
        print(no, element.text)
        # DataFrame에 리뷰 추가
        reviews_df.loc[len(reviews_df)] = [no, element.text]
    
    if no <= cnt and x < page_cnt:
        try:
            # 다음 페이지 이동 버튼을 누르기 위한 XPath 조정 필요
            next_button = driver.find_element(By.XPATH, '//*[@id="REVIEW"]/div/div[3]/div[2]/div/div/a[3]')

            next_button.click()
        except Exception as e:
            print(f'다음 페이지 이동 중 오류 발생: {e}')
            break
        time.sleep(2)

print('========= 작업 완료 =========')
```
